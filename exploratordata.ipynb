```python
Australiameanandstd = pd.DataFrame()
Omanmeanandstd = pd.DataFrame()
SaintLuciameanandstd = pd.DataFrame()
UnitedStatesmeanandstd = pd.DataFrame()
Ecuadormeanandstd = pd.DataFrame() 
```

```python
Australiameanandstd['mean'] = Australia.mean(axis=0).round(decimals=2)
Omanmeanandstd['mean'] = Oman.mean(axis=0).round(decimals=2)
SaintLuciameanandstd['mean'] = SaintLucia.mean(axis=0).round(decimals=2)
UnitedStatesmeanandstd['mean'] = UnitedStates.mean(axis=0).round(decimals=2)
Ecuadormeanandstd['mean'] = Ecuador.mean(axis=0).round(decimals=2)
```

```python
Australiameanandstd['standard deviation'] = Australia.std(axis=0)
Omanmeanandstd['standard deviation'] = Oman.std(axis=0) 
UnitedStatesmeanandstd['standard deviation'] = UnitedStates.std(axis=0) 
Ecuadormeanandstd['standard deviation'] = Ecuador.std(axis=0)
SaintLuciameanandstd['standard deviation'] = SaintLucia.std(axis=0)
```
